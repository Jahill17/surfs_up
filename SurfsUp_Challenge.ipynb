```python
# Dependencies
import numpy as np
import pandas as pd

%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt

import datetime as dt

# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)

# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

## D1: Determine the Summary Statistics for June

```python
# 1. Import the sqlalchemy extract function.
from sqlalchemy import extract

# 2. Write a query that filters the Measurement table to retrieve the temperatures for the month of June. 
june_temps = session.query(Measurement.date, Measurement.tobs).\
    filter(extract('month', Measurement.date)==6)
print(june_temps)
```

```python
#  3. Convert the June temperatures to a list.

june_temps = session.query(Measurement.date, Measurement.tobs).\
    filter(extract('month', Measurement.date)==6).all()
print(june_temps)
```

```python
# 4. Create a DataFrame from the list of temperatures for the month of June. 
june_temps_df = pd.DataFrame(june_temps, columns=['date', 'June Temps'])

#june_temps_df

print(june_temps_df.to_string(index=False))
```

```python
# 5. Calculate and print out the summary statistics for the June temperature DataFrame.

print(june_temps_df.describe().to_string(index=True))

# Format the dataframe
june_temps_df.describe()
```

## D2: Determine the Summary Statistics for December

```python
# 6. Write a query that filters the Measurement table to retrieve the temperatures for the month of December.

dec_temps = session.query(Measurement.date, Measurement.tobs).\
    filter(extract('month', Measurement.date)==12)
print(dec_temps)
```

```python
# 7. Convert the December temperatures to a list.

dec_temps = session.query(Measurement.date, Measurement.tobs).\
    filter(extract('month', Measurement.date)==12).all()
print(dec_temps)
```

```python
# 8. Create a DataFrame from the list of temperatures for the month of December. 

dec_temps_df = pd.DataFrame(dec_temps, columns=['date', 'Dec Temps'])
dec_temps_df
```

```python
# 9. Calculate and print out the summary statistics for the Decemeber temperature DataFrame.

print(dec_temps_df.describe().to_string(index=True))

# Format the dataframe
dec_temps_df.describe()
```

```python
# June station query

station_june = session.query(Measurement.tobs, Measurement.station).filter(extract('month',Measurement.date)== 6).all()
print(station_june)
```

```python
# June station query continue.  Create dataframe

station_june_df = pd.DataFrame(station_june, columns=['June Temps', 'Station'])
station_june_df
```

```python
# December station query

station_dec = session.query(Measurement.tobs, Measurement.station).filter(extract('month',Measurement.date)== 6).all()
print(station_dec)
```

```python
# December station query continue.  Create dataframe

station_dec_df = pd.DataFrame(station_dec, columns=['Dec Temps', 'Station'])
station_dec_df
```

```python
# Set up combined df merging stations

station_temps_combo = station_june_df.merge(station_dec_df, on="Station", how="inner").set_index("Station")
station_temps_combo
```
